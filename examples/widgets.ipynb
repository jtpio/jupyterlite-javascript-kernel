# Jupyter Widgets

The JavaScript kernel provides built-in widget classes that mirror the
[ipywidgets](https://ipywidgets.readthedocs.io/) API. Widget classes are
available under `Jupyter.widgets`.

Run the setup cell below first to destructure the widget classes used in this
notebook.

Widgets **auto-display** when they are the last expression in a cell.
You can also call the global `display()` function explicitly, e.g. when assigning to a variable.

> **Requirements:** `jupyterlab-widgets`, `@jupyter-widgets/controls`, and
> `@jupyter-widgets/output` must be available in the JupyterLite deployment.

```javascript
const {
    IntSlider,
    FloatSlider,
    FloatLogSlider,
    IntRangeSlider,
    FloatRangeSlider,
    SelectionSlider,
    SelectionRangeSlider,
    IntProgress,
    Play,
    IntText,
    FloatText,
    Text,
    Textarea,
    Password,
    Checkbox,
    ToggleButton,
    Valid,
    Dropdown,
    RadioButtons,
    Select,
    SelectMultiple,
    ToggleButtons,
    Button,
    ButtonStyle,
    HTML,
    Label,
    Output,
    Layout,
    ColorPicker,
    VBox,
    HBox,
    Tab,
    Accordion,
    Stack,
    jslink
} = Jupyter.widgets;
```

```javascript
const slider = new IntSlider({
    value: 50,
    min: 0,
    max: 100,
    step: 1,
    description: 'Int Slider'
});
display(slider);

slider.observe(({ new: value }) => {
    console.log('Slider value:', value);
}, 'value');
```

## Progress Bars

```javascript
const progress = new IntProgress({
    value: 0,
    min: 0,
    max: 100,
    description: 'Loading',
    bar_style: 'info'
});
display(progress);

// Animate the progress bar
const interval = setInterval(() => {
    progress.value += 5;
    if (progress.value >= 100) {
        clearInterval(interval);
        progress.bar_style = 'success';
        progress.description = 'Done';
    }
}, 200);
```

## Button

Buttons fire `'click'` events instead of holding a value:

```javascript
let clickCount = 0;

const btn = new Button({
    description: 'Click me!',
    button_style: 'primary',
    tooltip: 'Go ahead, click it'
});
display(btn);

btn.onClick(() => {
    clickCount++;
    console.log(`Clicked ${clickCount} time(s)`);
});
```

## Linking Widgets

Use the frontend link helpers when you want widgets to stay in sync without manual wiring:

```javascript
const play = new Play({ value: 20, min: 0, max: 100, interval: 150 });
const source = new IntSlider({ value: 20, min: 0, max: 100, description: 'Value' });
const mirror = new IntProgress({ value: 20, min: 0, max: 100, description: 'Mirror' });
const label  = new Label({ value: 'Value: 20' });

jslink([play, 'value'], [source, 'value']);
jslink([source, 'value'], [mirror, 'value']);
source.observe(({ new: value }) => {
    label.value = `Value: ${value}`;
}, 'value');

new VBox({
    children: [
        new HBox({ children: [play, source] }),
        mirror,
        label
    ]
})
```

## Layout, Style, and Output

Layout models, style models, and output widgets are regular widgets too.

```javascript
new Button({
    description: 'Styled button',
    button_style: 'info',
    layout: new Layout({
        width: '220px',
        justify_content: 'center'
    }),
    style: new ButtonStyle({
        button_color: '#ffd166',
        text_color: '#1f2937'
    })
})
```

```javascript
const out = new Output({
    layout: new Layout({
        border: '1px solid #d1d5db',
        padding: '8px',
        max_height: '140px',
        overflow: 'auto'
    })
});
display(out);

const logToOutput = out.capture({ clearOutput: true })(() => {
    console.log('Captured stdout inside Output.');
    out.appendDisplayData({
        'text/markdown': '**Direct append** from JavaScript',
        'text/plain': 'Direct append from JavaScript'
    });
});
logToOutput();
```

## Container Widgets

Container widgets group other widgets together and control their layout.
They accept a `children` array of widget instances.

```javascript
// VBox arranges children vertically
new VBox({
    children: [
        new IntSlider({ value: 30, description: 'Slider A' }),
        new IntSlider({ value: 60, description: 'Slider B' }),
        new IntSlider({ value: 90, description: 'Slider C' })
    ]
})
```

```javascript
// HBox arranges children horizontally
new HBox({
    children: [
        new Button({ description: 'Left', button_style: 'info' }),
        new Button({ description: 'Center', button_style: 'warning' }),
        new Button({ description: 'Right', button_style: 'danger' })
    ]
})
```

```javascript
// Tab shows one child at a time with tab headers
new Tab({
    children: [
        new IntSlider({ value: 50, description: 'Slider' }),
        new Text({ value: 'Hello', description: 'Name' }),
        new ColorPicker({ value: '#e74c3c', description: 'Color' })
    ],
    titles: ['Slider', 'Text', 'Color']
})
```

```javascript
// Accordion shows one child at a time with collapsible sections
new Accordion({
    children: [
        new VBox({
            children: [
                new IntSlider({ description: 'X' }),
                new IntSlider({ description: 'Y' })
            ]
        }),
        new HBox({
            children: [
                new ToggleButton({ description: 'A', button_style: 'success' }),
                new ToggleButton({ description: 'B', button_style: 'info' })
            ]
        })
    ],
    titles: ['Sliders', 'Toggles']
})
```

```javascript
// Stack displays only the selected child (controlled programmatically)
const stack = new Stack({
    children: [
        new HTML({ value: '<h3>Page 1</h3><p>First page content</p>' }),
        new HTML({ value: '<h3>Page 2</h3><p>Second page content</p>' }),
        new HTML({ value: '<h3>Page 3</h3><p>Third page content</p>' })
    ],
    titles: ['Page 1', 'Page 2', 'Page 3'],
    selected_index: 0
});
display(stack);

// Switch pages with a dropdown
const pagePicker = new Dropdown({
    options: ['Page 1', 'Page 2', 'Page 3'],
    description: 'Show page'
});
display(pagePicker);

pagePicker.on('change:index', () => {
    stack.selected_index = pagePicker.index;
});
```

## Low-level Comm API

The widget classes above are built on `Jupyter.comm`, the raw comm protocol
API. You can use it directly for custom frontend↔kernel communication.

> Custom target names that have no frontend handler will show
> `"Exception opening new comm"` in the browser console — this is expected.

```javascript
// Open a raw comm channel
const comm = Jupyter.comm.open('my.custom.target', { greeting: 'hello' });
console.log('Comm ID:', comm.commId);

// Send / receive / close
comm.send({ method: 'update', state: { value: 42 } });
comm.onMsg = (data) => console.log('Received:', data);
comm.close();
```

```javascript
// Register a target for frontend-initiated comms
Jupyter.comm.registerTarget('my.echo', (comm, data) => {
    comm.onMsg = (msg) => comm.send({ echo: msg });
});
```
